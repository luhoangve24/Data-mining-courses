**Khai báo thư viện**

```python
import pandas as pd
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression 
```

**Đọc dữ liệu**

```python
df = pd.read_csv("bank_data1.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
#Hiển thị kiểu dữ liệu
df.dtypes
```

```python
#Đổi dữ liệu từ định danh (ofject) về dạng số
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
df["sex"] = lb_make.fit_transform(df["sex"])
df["region"] = lb_make.fit_transform(df["region"])
df["married"] = lb_make.fit_transform(df["married"])
df["car"] = lb_make.fit_transform(df["car"])
df["save_act"] = lb_make.fit_transform(df["save_act"])
df["current_act"] = lb_make.fit_transform(df["current_act"])
df["mortgage"] = lb_make.fit_transform(df["mortgage"])
df["pep"] = lb_make.fit_transform(df["pep"])
df.head(10)
```

```python
# Xác định thuộc tính mô tả X và thuộc tính dự đoán y
features = ['age','sex','region','income','married','children','car',
'save_act','current_act','mortgage']
target = ['pep']
X = df[features]
y = df[target]
print(X)
print(y)
```

```python
# Chia bộ dữ liệu thành hai tập train và test theo tỉ lệ 80% train,20% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
# Khai báo mô hình cây quyết định
model=tree.DecisionTreeClassifier(criterion = "entropy",
                                  random_state = 100,
                                  max_depth = 3,
                                  min_samples_leaf = 5)
```

```python
# Thực thi mô hình
model = model.fit(X_train,y_train)
```

```python
 # Kiểm thử mô hình
y_pred = model.predict(X_test)
print("Giá trị nhãn mô hình dự đoán được:\n " ,y_pred)
```

```python
# Xác định ma trần nhầm lẫn
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
```

```python
# Độ chính xác của mô hình
print ("Độ chính xác: ", accuracy_score(y_test, y_pred)*100)
```

```python
# Sử dụng mô hình dự đoán khả năng trả nợ của khách hàng có các thông tin sau
# Tuoi 42, Gioi_Tinh nữ 1, Khu_Vuc thị trấn 3, Thu_Nhap 30527, có Ket_Hon 1
# So_Con 2, có O_To 1, có TK_Tiet_Kiem 1, không có TK_Thanh_Toan 0, có The_Cha
#p 1

x = [[42,1,3,30527,1,2,1,1,0,1]]
y = model.predict(x)
if y==1:
    print ("Khách hàng trả được nợ")
else:
    print("Khách hàng không trả được nợ")
```

**Bài tập**

Dự đoán tuổi của khách hàng dựa vào giới tính, thu nhập, số con và khả năng trả nợ

```python
#Xác định thuộc tính mô tả thuộc tính X và Y
features = ['sex','income','children','pep','mortgage'] 
# khả năng trả nợ hay PEP, mortgage là có thế chấp hay không
# đoạn này cần lưu ý các thuộc tính mô tả yêu cầu là gì -> bỏ mortgage đi nhé, hoặc
# Nếu dự đoán như ở cuối cùng, thì bài toán có thể sửa như sau
# 1, Bỏ thuộc tính mô tả mortgage, thêm thuộc tính region phía sau sex
# 2, Thêm mortgage thì phần dự đoán bỏ region đi, thêm số liệu cho thuộc tính mortgage ở phía cuối, sau pep
# Vì train mortgage ở đầu rồi thì làm theo bước 2 --->
target = ['age']
X = df[features]
y= df[target]
print(X)
print(y)
```

```python
#Chia bộ dữ liệu thành 2 tập train và test 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
#Khai báo mô hình hồi quy tuyến tính
model = LinearRegression()
```

```python
# Thực thi mô hình
model = model.fit(X_train,y_train)
```

```python
print("Mô hình hồi quy sẽ có dạng: \n")
print("age = b+ a0*sex + a1*region + a1*income + a3*children + a4*pep")
print("Với các hệ số lần lượt là: ")
print(model.intercept_)
print(model.coef_)
```

```python
#Kiểm thử mô hình
y_pred = model.predict(X_test)
print(y_pred)
```

```python
#Độ chính xác của mô hình
f = model.score(X_test, y_test)
print(f)
```

```python
#Sử dụng mô hình
# sex, region, income, children, pep --> xác định sai thuộc tính
# x = [[1, 3, 30333.9, 3, 1]]
# thuộc tính đúng: sex, income, children, pep, mortgage --> theo hướng 1
x = [[1, 30333.9, 3, 1, 1]] # --> thêm dữ liệu cho mortgage (1 hoặc 0)
y = model.predict(x)
print("Tuổi của khách hàng là:",y)
```
