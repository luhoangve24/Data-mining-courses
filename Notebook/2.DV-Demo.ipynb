```python
# Phương thức subplots trả về cả Figure và Axes nên có thể dùng
# thuộc tính phương thức của cả hai đối tượng này chung trong phương thức
# Thuộc tính: số dòng, số cột, kích cỡ Figure
# Xem thêm: ở trên
# figsize: width và heigth trong inches
# dpi: độ phân giải dots-per-inches
# sharey/sharex: True để cùng các ticks label của 2 trục
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(10,4), sharey=True, dpi=80)
ax1.bar(df_ketoan['SoNamKinhNghiem'],df_ketoan['Luong'])
ax2.bar(df_hcns['SoNamKinhNghiem'],df_hcns['Luong'])
ax3.bar(df_sale['SoNamKinhNghiem'],df_sale['Luong'])
# Tieu de bieu do
ax1.set_title('Ke toan')
ax2.set_title('Hanh chinh nhan su')
ax3.set_title('Sale')

# Tieu de truc hoanh
ax1.set_xlabel('So nam kinh nghiem')
ax2.set_xlabel('So nam kinh nghiem')
ax3.set_xlabel('So nam kinh nghiem')

# Tieu de truc tung
ax1.set_ylabel('Luong')
ax2.set_ylabel('Luong')
ax3.set_ylabel('Luong')

# Dieu chinh khoang cach giua cac subplots
plt.tight_layout()
plt.show()
```

##### Biểu đồ thanh xếp chồng - Stacked Bar Chart
(Thể hiện trung bình cộng lương và trung bình cộng số năm kinh nghiệm theo nhóm nghề)\
[Xem thêm: Groupby Tricks để hiểu](https://builtin.com/data-science/pandas-groupby)\
[Stacked Bar Chart: Official Guide](https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_stacked.html#sphx-glr-gallery-lines-bars-and-markers-bar-stacked-py)

```python
tbc_luong_nghe = df.groupby('NganhNghe')[['Luong']].mean()
tbc_nam_nghe = df.groupby('NganhNghe')[['SoNamKinhNghiem']].mean()
print('TBC Luong theo Nghe\n', tbc_luong_nghe, '\n')
print('TBC NKN theo Nghe\n', tbc_nam_nghe, '\n')

plt.xlabel('Cac nganh nghe', color='green', fontweight='bold', fontsize='15')

# pyplot.bar(x, height, ...)
plt.bar(tbc_luong_nghe.index, tbc_luong_nghe['Luong'], label='TBC Luong')
plt.bar(tbc_nam_nghe.index, tbc_nam_nghe['SoNamKinhNghiem'], label='TBC Nam kinh nghiem')

plt.legend()
plt.show()
```

##### Biểu đồ thanh nhóm - Grouped Bar Chart
(Min, Max, Mean theo Lương)\
[Grouped Bar Chart: Official Guide](https://matplotlib.org/stable/gallery/lines_bars_and_markers/barchart.html#sphx-glr-gallery-lines-bars-and-markers-barchart-py)

```python
max_luong_nghe = df.groupby('NganhNghe')[['Luong']].max()
min_luong_nghe = df.groupby('NganhNghe')[['Luong']].min()
tbc_luong_nghe = df.groupby('NganhNghe')[['Luong']].mean()

# The label location (Vi tri hien thi cac nhom)
x = np.arange(len(max_luong_nghe.index))
width = 0.2 # Kich thuoc cac thanh

# array in numpy can sum one number to every numbers
fig, ax = plt.subplots()
ax.bar(x, max_luong_nghe['Luong'], width, label='Max Luong')
ax.bar(x+0.2, min_luong_nghe['Luong'], width, label='Min Luong')
ax.bar(x+0.4, tbc_luong_nghe['Luong'], width, label='TBC Luong')

ax.set_title('Gia tri luong theo nganh nghe')
ax.set_xticks(x)
ax.set_xticklabels(max_luong_nghe.index)
ax.legend()

fig.tight_layout()

plt.show()
```

##### Biểu đồ phân tán Scatter
(Miêu tả sự phân bố của dữ liệu trong không gian hai chiều -> Tìm ra mối tương quan)\
[Đọc thêm: guide](https://saturncloud.io/blog/how-to-add-a-regression-line-in-python-using-matplotlib/)\
[Đọc thêm: guide 2](https://www.statology.org/scatterplot-with-regression-line-python/)

```python
height = np.random.randint(140, 200, size=15)
weight = np.random.randint(65, 100, size=15)
colors = np.random.rand(15)
area = (30 * np.random.rand(15))**2

plt.xlim(130,220)
plt.ylim(60,120)
plt.scatter(height, weight, s=area, c=colors)
plt.title('Chieu cao va can nang')
plt.xlabel('Chieu cao')
plt.ylabel('Can nang')

# Lam them: Regression Linear Line
# m: slope (do doc), b: intercept (he so chan)
m, b = np.polyfit(height, weight, 1)
plt.plot(height, m*height+b)

plt.show()
```
