### Khai báo thư viện

```python
import pandas as pd
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
```

### Đọc dữ liệu

```python
df = pd.read_csv('../Data/bank_data.csv')
```

### Hiểu dữ liệu

```python
df.head(5) # Xem 5 dòng đầu
```

```python
df.tail(5) # Xem 5 dòng cuối
```

```python
# Hiển thị kiểu dữ liệu
df.dtypes
```

```python
# Đổi dữ liệu từ định danh (object) về dạng số
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
df['sex'] = lb_make.fit_transform(df['sex'])
df['region'] = lb_make.fit_transform(df['region'])
df['married'] = lb_make.fit_transform(df['married'])
df['car'] = lb_make.fit_transform(df['car'])
df['save_act'] = lb_make.fit_transform(df['save_act'])
df['current_act'] = lb_make.fit_transform(df['current_act'])
df['mortgage'] = lb_make.fit_transform(df['mortgage'])
df['pep'] = lb_make.fit_transform(df['pep'])
df.head(10) # Kiểm tra lại 10 dòng dữ liệu đầu
```

```python
# Xác định thuộc tính mô tả X và thuộc tính dự đoán Y
features = ['age','sex','region','income','married','children','car','save_act','current_act','mortgage']
target = ['pep']
X = df[features]
y = df[target]
print(X)
print(y)
```

```python
# Chia bộ dữ liệu thành tập train và test, random_state mặc định = 0 (lấy linh tinh)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
# Khai báo mô hình cây quyết định
model = tree.DecisionTreeClassifier(criterion = 'entropy',
                                   random_state = 100,
                                   max_depth = 7,
                                   min_samples_leaf = 5)
```

```python
# Thực thi mô hình
model = model.fit(X_train, y_train)
```

### Xác định độ chính xác mô hình

```python
# Kiểm thử mô hình
y_pred = model.predict(X_test)
print("Giá trị nhãn mô hình dự đoán được: \n", y_pred)
```

```python
# Xác định ma trận nhầm lẫn
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
```

```python
# Độ chính xác của mô hình
print("Độ chính xác: ", accuracy_score(y_test, y_pred)*100,'%')
```

```python
# Sử dụng mô hình
x=[[42,1,3,30527,1,2,1,1,0,1]]
y=model.predict(x)
if y==1:
    print("Khách trả nợ được")
else:
    print("Khách không trả nợ")
```

### Bài tập
Dự đoán tuổi của khách hàng dựa vào giới tính, khu vực, thu nhập, số con & khả năng trả nợ

```python
# Xác định thuộc tính mô tả X và thuộc tính dự đoán Y
features = ['sex','region','income','children','pep']
target = ['age']
X = df[features]
y = df[target]
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
# Khai báo mô hình Hồi quy tuyến tính
model = LinearRegression()
```

```python
# Thực thi mô hình
model = model.fit(X_train, y_train)
```

```python
print("Mô hình hồi quy sẽ có dạng:")
print("age = a0 + a1*sex + a2*region + a3*income + a4*children + a5*pep")
print("Với các hệ số lần lượt là:")
print(model.intercept_)
print(model.coef_)
```

```python
# Kiểm thử mô hình
y_pred = model.predict(X_test)
print(y_pred)
```

```python
# Độ chính xác của mô hình
f = model.score(X_test, y_test)
print(f)
```
