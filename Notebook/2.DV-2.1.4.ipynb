```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Data/FoodPrice_in_Turkey.csv')
```

```python
print(df.info())
```

[How to Filter data in Dataframe?](https://builtin.com/data-science/pandas-filter)

```python
# Muc tieu
# (1) Bar Chart: Gia gao thang 12/2019 cua Ankara, Istanbul, Izmir va National Average
# (2) Line Chart: Xu huong gia gao trung binh ca nuoc (NA) 2019 tai Tho Nhi Ki
# (3) Scatter Plot: Correlation  ~ Gia gao va Gia gas NA tai Tho Nhi Ki
# --- Come in to play (1)! ---

df_1 = df[(df['Year'] == 2019) & (df['Month'] == 12) & (df['ProductName'] == 'Rice - Retail')]
```

```python
(df['Place'].value_counts()).index
```

```python
plt.bar(df_1['Place'], df_1['Price'], width=0.5, label=(df['Place'].value_counts()).index, color=(['r','g','b', 'pink']))
plt.title('National Rice Price in 12/2019 of Turkey', fontsize=16, color='r')
plt.xlabel('Place', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.legend()
plt.show()
```

```python
# --- (2) ---
df_2 = df[(df.Place == 'National Average') & (df.Year == 2019) & (df.ProductName == 'Rice - Retail')]
```

```python
plt.plot(df_2['Month'], df_2['Price'], linewidth=2, marker='*', markersize=10, markerfacecolor='red', markeredgecolor='blue', markeredgewidth=2)
plt.title('Rice Price of National Average in 2019', fontsize=16, color='r')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.show()
```

```python
# --- (3) ---
x = df[(df.Place == 'National Average') & (df.ProductName == 'Fuel (gas) - Retail') & (df.Year == 2019)]
y = df[(df.Place == 'National Average') & (df.ProductName == 'Rice - Retail') & (df.Year == 2019)]
```

```python
import numpy as np
plt.scatter(x.Price, y.Price, s=50)
plt.title('Relationship between Rice Price and Gas Price', fontsize=16, color='r')
plt.xlabel('Gas', fontsize=14)
plt.ylabel('Rice', fontsize=14)
s, i = np.polyfit(x.Price, y.Price, 1)
plt.plot(x.Price, s*x.Price+i)
plt.show()
```

```python
# --- (3) --- (Remove Outliners)
x = df[(df.Place == 'National Average') & (df.ProductName == 'Fuel (gas) - Retail') & (df.Year == 2019) & (df.index != 1927)]
y = df[(df.Place == 'National Average') & (df.ProductName == 'Rice - Retail') & (df.Year == 2019) & (df.index != 64)]
```

```python
import numpy as np
plt.scatter(x.Price, y.Price, s=50)
plt.title('Relationship between Rice Price and Gas Price', fontsize=16, color='r')
plt.xlabel('Gas', fontsize=14)
plt.ylabel('Rice', fontsize=14)
s, i = np.polyfit(x.Price, y.Price, 1)
plt.plot(x.Price, s*x.Price+i)
plt.show()
```
