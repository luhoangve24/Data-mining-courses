```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Data/daily-min-temperatures.csv')
```

```python
df.info()
df.head(10)
```

```python
df.Date = pd.to_datetime(df.Date)
```

```python
df.Temp.describe()
```

```python
plt.hist(df.Temp, bins=45, range=(1, 23), histtype='step')
plt.title('Temperature Histogram', fontsize=16)
plt.show()
```

```python
plt.plot(df.Date, df.Temp)
plt.title('Daily Temperature', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Temp', fontsize=14)
plt.show()
```

```python
bounds = ['1/1/1990', '12/31/1990']
bounds = pd.to_datetime(bounds)
df_custom = df[(df.Date >= bounds[0]) & (df.Date <= bounds[1])]
plt.plot(df_custom.Date, df_custom.Temp)
plt.title('Daily Temp in 1990', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Temp', fontsize=14)
plt.show()
```
